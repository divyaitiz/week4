```python
import numpy as np
import pandas as pd
```

```python
number_of_samples = 100
```

```python
np.random.seed(0)
```

Assuming a linear relationship:

y = w1x1 + w2X2 + w3X3 + e(noise)

```python
x1 = np.random.uniform(0, 10, number_of_samples)
x2 = np.random.uniform(-5, 5,number_of_samples)
x3 = np.random.uniform(0, 1, number_of_samples)
```

```python
print(x1)
```

```python
x2
```

```python
x3
```

```python
epsilon = np.random.normal(0,1,number_of_samples)
```

```python
y = 4*x1 - 1.5*x2 + 2*x3 + epsilon

# Create DataFrame
regression_data = pd.DataFrame({
    'X1': x1,
    'X2': x2,
    'X3': x3,
    'target': y
})

print(regression_data.head())
```

```python
X = np.column_stack((x1, x2, x3))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted (Linear Regression)")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("MSE:", mse)
print("R^2:", r2)
```
